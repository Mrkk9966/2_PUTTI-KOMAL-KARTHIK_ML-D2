```python
df= pd.read_csv('passenger_list.csv')
```

```python
df.head()
```

```python
print(df.columns)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.scatter(df["Age"], df["Seat_Number"])

plt.xlabel("Age")
plt.ylabel("Ticket Price")
plt.title("Age vs Ticket Price Scatter Plot")

plt.show()
```

```python
sns.barplot(x='Gender', y='Age', data=df)
plt.title("Average Age by Gender")
```

```python
sns.boxplot(x='Gender', y='Age', data=df)
plt.title("Age Distribution by Gender")
```

```python
sns.distplot(df['Age'])
plt.title("Age Distribution by Gender")
plt.show()
```

```python
df.loc[5, 'Age'] = None
```

```python
sns.heatmap(df.isnull(), cbar=False)

plt.title("Missing Values Heatmap")
plt.show()
```

```python
corr = df.corr(numeric_only=True)

sns.clustermap(corr, annot=True, cmap='coolwarm')

plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
sns.lineplot(x='Age', y='Ticket_Type', data=df)

plt.title("Age vs Ticket Price")
```
